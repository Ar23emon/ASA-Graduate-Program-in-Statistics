```python
columns_to_count = ['PhD', 'MASTER\'S', 'Data Science UG', 'Data Science Master\'s', 'Data Science PhD', 'UG MAJOR', 'UG CONCENTRATION', 'UG MINOR']

count_dict = {}

for column_name in columns_to_count:
    count_ones = df[column_name].eq('1').sum()
    count_dict[column_name] = count_ones

plt.figure(figsize=(10, 6))
sns.barplot(x=list(count_dict.keys()), y=list(count_dict.values()))
plt.xlabel('Program')
plt.ylabel('Number Of Total Instituition')
plt.title('Programs offer by All Instituition(out of 201)')
plt.xticks(rotation=45)
plt.show()
```

## The Statistical Departments offers maximum number of program type in:
#### 1) Undergraduate Major in Statistics
#### 2) Masters in Statistics
#### 3) Undergraduate Minor in Statistics
#### 4) PhD in Statistics
#### 5) Undergraduate in Data Science
#### 6) Masters in Data Science
#### 7) PhD in Data Science
